#### Contexte
L'ensemble de données comprend des annonces provenant de portails immobiliers populaires de Madrid 

#### Emplacement 
Madrid est l'une des villes les plus visitées d'Europe, tant par les touristes que par les hommes d'affaires, et c'est là que de nombreuses entreprises locales et multinationales importantes ont leur siège social. Par conséquent, Madrid bénéficie à la fois d'un afflux important de touristes et de personnes cherchant à vivre et à travailler dans la ville pour donner un coup de pouce à leur carrière professionnelle.

#### Prix attractifs 
Le marché immobilier est en phase de croissance depuis quelques années maintenant, ce qui signifie que vous pouvez toujours trouver des propriétés à des prix très attractifs (surtout par rapport aux autres capitales européennes) mais avec encore de bonnes chances de voir leur valeur augmenter à l'avenir.

#### Types de maisons
Les maisons individuelles ne sont pas courantes dans le centre-ville, mais existent dans les banlieues. Madrid propose également un mélange de nouvelles constructions et d'appartements existants. Les prix de l'immobilier dans les capitales sont généralement les plus chers d'un pays, mais ce n'est pas le cas en Espagne. Le coût de l'immobilier à Madrid est bien inférieur à celui de Barcelone et des zones côtières.
