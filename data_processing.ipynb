##### Description du nettoyage effectué sur les données

Ici, nous allons supprimer les colonnes comportant les valeurs manquantes a plus de 80% : 
- n_floors : nombre d'étages 
- sq_mt_allotment : m carré attribué
- latiude 
- longitude 
- portal : présence d'un portail 
- door : nombre de porte 
- are_pets_allowed : animaux autorisés 
- is_furnished : meublé
- is_kitchen_equipped : cuisine équipé
- is_accessible
- has_green_zones           
- has_private_parking : contient un parking privé 
- has_public_parking : contient un parking publique 
Certaines de ces données peuvent avoir un réel impact sur le prix comme par exemple le nombre d'étage, la présence d'un portail, la présence de meubles ou de cuisine équipé mais le nombre de valeurs manquantes est beaucoup trop élevé 

On supprime la colonne "title" qui n'a pas d'intérêt et d'impact sur la target du prix.

Nous supprimons la colonne "sq_mt_useful", qui indique les mètres carrés utilisés, car elle n'est pas utile pour la prédiction. En effet, la colonne "sq_mt_built", qui représente les mètres carrés construits, influe directement sur le prix.

Nous supprimons la colonne "is_buy_price", qui indique si le prix du bien est renseigné, car la colonne "buy_price" nous donne déjà cette information : si elle est vide, cela signifie que le prix n'est pas renseigné.

Nous supprimons la colonne "has_garden", qui contient plus de 92 % de valeurs manquantes, et parce que nous disposons d'une autre variable similaire, "has_exterior".

Nous supprimons les colonnes "is_parking_include_in_price" et "parking_price", qui indiquent respectivement si le prix du parking est inclus dans le prix d'achat et le prix du parking, car plus de 64 % des données sont manquantes pour ces deux colonnes.

On supprime également les colonnes concernant l'adresse et le lieu : "subtitle", "is_exact_address_hidden", "street_name", "street_number", "raw_adress", "neighboorhod_id", puisqu'on va seulement récupèrer le prix au mètre carré situé dans "neighboorhod_id" et le stocker dans une nouvelle variable "sq_mt_price_by_place".

On supprime "is_floor_under" qui comporte seulement deux valeurs uniques "null" et " ".

On supprime "operation" qui contient qu'une valeur unique "sale".

Nous supprimons "is_rent_price_known" puisque il y a qu'une seule valeur unique "False".




Pour les variables "subtitle" et "house_type_id", on supprime les itérations qui reviennent et sont inutiles.

Pour la variable "floor" on remplace les chaînes de caractères par la valeur null et on transforme le type en int.
On transforme également le type de la colonne "n_bathrooms" en int.

On remplace les valeurs null par "False" lorsque les colonnes comporte les valeurs uniques "True" et null.

On remplace les valeurs null par "None" si le type des valeurs est object.

On remplace les valeurs null de la colonne "n_rooms" par la médiane. 

```python
def data_cleaned_madrid(df):
    # Supression des colonnes 
    df.drop(columns=['Unnamed: 0'], inplace=True)
    df.set_index('id', inplace=True)
    missing_percentages = df.isnull().mean() * 100
    cols_to_drop = missing_percentages[missing_percentages > 80].index
    df.drop(columns=cols_to_drop, inplace=True)
    df.drop(columns=['title', 'subtitle', 'sq_mt_useful', 'is_buy_price_known','is_parking_included_in_price', 'parking_price', 'raw_address', 'is_exact_address_hidden', 'street_name', 'street_number', 'is_floor_under', 'operation', 'is_rent_price_known'], inplace=True)
    
    # Suppression des itérations pour les variables qualitatives
    if 'subtitle' in df:
        df['subtitle'] = df['subtitle'].str.replace(', Madrid', '', regex=False)
    if 'house_type_id' in df.columns:
        df['house_type_id'] = df['house_type_id'].str.split(': ', n=1).str[1]
    
    df['sq_mt_price_by_place'] = df['neighborhood_id'].str.extract(r'(\d+\.?\d*) €/m2')
    df['sq_mt_price_by_place'] = df['sq_mt_price_by_place'].astype(float)

    df.drop(columns=['neighborhood_id'], inplace=True)

    # Supprimer les valeurs textuelles dans la colonne "floor"
    df['floor'] = df['floor'].str.extract(r'(\d+)')

    # Transformation des types 
    df['floor'] = df['floor'].astype('Int64')
    df['n_bathrooms'] = df['n_bathrooms'].astype('Int64')

    # Remplacer les valeurs null par False lorsque les valeurs uniques sont True ou null
    df['has_ac'] = df['has_ac'].fillna(False)
    df['has_fitted_wardrobes'] = df['has_fitted_wardrobes'].fillna(False)
    df['has_pool'] = df['has_pool'].fillna(False)
    df['has_terrace'] = df['has_terrace'].fillna(False)
    df['has_storage_room'] = df['has_storage_room'].fillna(False)

    # Remplacer les valeurs null par 'None' si le type des valeurs est object
    # object_columns = df.select_dtypes(include=['object']).columns
    # df[object_columns] = df[object_columns].fillna('None')

    # Imputation des valeurs manquantes
    median_n_rooms = df['n_rooms'].median()
    df['n_rooms'].fillna(median_n_rooms, inplace=True)
    median_n_rooms = df['n_bathrooms'].median()
    df['n_bathrooms'].fillna(median_n_rooms, inplace=True)

    return df

df_cleaned = data_cleaned_madrid(df)
```

#### Détection des outliers 

```python
variables = ['sq_mt_built', 'buy_price', 'sq_mt_price_by_place']

plt.figure(figsize=(15, 5))

for i, var in enumerate(variables, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df_cleaned[var])
    plt.title(f'Boxplot de {var}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
def remove_outliers(df, column):
    print(f'Nombre de lignes avant filtrage pour {column}: {len(df)}')
    
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    print(f'Nombre de lignes après filtrage pour {column}: {len(df_filtered)}')
    
    return df_filtered

df_without_outliers = remove_outliers(df_cleaned, 'sq_mt_built')
df_without_outliers = remove_outliers(df_without_outliers, 'buy_price')
df_without_outliers = remove_outliers(df_without_outliers, 'sq_mt_price_by_place')
```

```python
variables = ['sq_mt_built', 'buy_price', 'sq_mt_price_by_place']

plt.figure(figsize=(15, 5))

for i, var in enumerate(variables, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=df_without_outliers[var])
    plt.title(f'Boxplot de {var}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```
